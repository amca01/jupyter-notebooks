```python
# import needed libraries and functions

import folium
import pandas as pd
from folium.plugins import BeautifyIcon
from folium.features import DivIcon
import geopandas as gpd
import shapely as sh
import numpy as np
from math import floor, ceil

import matplotlib.pyplot as plt
import matplotlib.colors as mcls
```

```python
vic_state = gpd.read_file("Vic.Maps/STATE_ASSEMBLY_2022.shp")  
```

```python
brunswick = vic_state[vic_state["DISTRICT"]=="BRUNSWICK"]
brunswick_crs=brunswick.to_crs(epsg=4326)
bounds = brunswick_crs.geometry.total_bounds
bounds
```

```python
brunswick.geometry.crs
```

```python
brunswick_json= brunswick_crs.to_json()
brunswick_j = folium.GeoJson(data=brunswick_json)
```

```python
extent = 0.1
[b0,b1,b2,b3] = bounds
centre = [(b1+b3)/2,(b0+b2)/2]

bmap = folium.Map(location=centre, crs='EPSG3857', tiles='OpenStreetMap',
                 min_lat=b1-extent, max_lat=b3+extent,
                 min_long=b0-extent, max_long=b2+extent,
               width=800,height=800,zoom_start=13,scrollWheelZoom=False)

brunswick_j.add_to(bmap)
folium.Rectangle([(b1,b0),(b3,b2)]).add_to(bmap)
bmap
```
